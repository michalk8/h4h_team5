```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pickle
from collections import defaultdict
```

```python
data = defaultdict(dict)
for size in [25, 50, 400]:
    for quality in (5, 20, 40, 60, 80, 95):
        with open(f'/home/michal/h4h_team5/model/results/result_{size}_jpeg_{quality}.pickle', 'rb') as fin:
            data[size][quality] = pickle.load(fin)
```

```python
for d in data.keys():
    for k in data[d].keys():
        for t in ('train', 'val'):
            tmp = []
            for i, cmat in enumerate(data[d][k]['cmats'][t]):
                tmp.append(np.diag(cmat).sum() / np.sum(cmat))
            data[d][k]['accs'][t] = tmp
```

```python
data[d][k].keys()
```

```python
fig, axes = plt.subplots(1, 3, figsize=(30, 10), sharey=True)

for i, d in enumerate(data.keys()):
    x = []
    ys = []
    for k in data[d].keys():
        x.append(k)
        ys.append([#data[d][k]['accs']['val'][-1],
                   data[d][k]['prec_avg']['val'][-1],
                   data[d][k]['rec_avg']['val'][-1],
                   data[d][k]['f1_avg']['val'][-1]])
    ax = axes[i]
    ys = np.array(ys).T    
    for name, y in zip(['Precision', 'Recall', 'F1'], ys):
        ax.plot(x, y, label=name)
    ax.set_xticks([5, 20, 40, 60, 80, 95])
    ax.set_xlabel('JPEG Quality')
    ax.set_ylabel('Accuracy')
    ax.set_title(f'Size @ {d}x{d}px')
    ax.legend()
fig.savefig('quality.png', bbox_inches='tight')
```
